### Summary Statistics Interpretation

The dataset contains 244 complete observations for numerical variables.

The average total bill is approximately $19.79, while the average tip is $2.99. 
The maximum total bill recorded is $50.81, and the highest tip given is $10.00.

The median tip (50th percentile) is $2.90, which is close to the mean, suggesting that tipping behavior is relatively consistent without extreme skewness.

Table sizes range from 1 to 6 people, with most tables appearing to have around 2 to 3 guests based on the quartile distribution.

Overall, the dataset shows realistic restaurant transaction behavior with moderate variability in spending and tipping.
